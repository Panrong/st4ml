```python
import osmnx as ox
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
%%time
G = ox.graph_from_bbox(41.14570, 41.14166, -8.61174, -8.62016, network_type='drive', simplify=True, retain_all=True)
print('G_drive:', len(G.nodes()), 'nodes', len(G.edges()), 'edges')
ox.plot_graph(G, node_color='b', node_size=3, edge_color='k', edge_linewidth=0.25, bgcolor='w', figsize=(12,12))
```

```python
print(type(G))
ox.save_graphml(G, filepath='test.graphml')
```
