```python
import osmnx as ox
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
%%time
ox.config(all_oneway=False)
G_drive = ox.graph_from_bbox(41.25, 41.10, -8.50, -8.70, network_type='drive', simplify=False, retain_all=True)
G_drive_simplified = ox.graph_from_bbox(41.25, 41.10, -8.50, -8.70, network_type='drive', simplify=True, retain_all=True)
```

```python
type(G_drive)
```

```python
print('G_drive:', len(G_drive.nodes()), 'nodes', len(G_drive.edges()), 'edges')
print('G_drive_simplified:', len(G_drive_simplified.nodes()), 'nodes', len(G_drive_simplified.edges()), 'edges')
```

```python
ox.plot_graph(G_drive, node_color='b', node_size=3, edge_color='k', edge_linewidth=0.25, bgcolor='w', figsize=(12,12))
```

```python
ox.plot_graph(G_drive_simplified, node_color='b', node_size=3, edge_color='k', edge_linewidth=0.25, bgcolor='w', figsize=(12,12))
```

```python
ox.io.save_graphml(G_drive_simplified, filepath='porto.graphml')
```
